```python
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('test_2^28_avx2.csv', delimiter=',')
data_o3 = np.genfromtxt('test_2^28_avx2_o3.csv', delimiter=',')
data_baseline = np.genfromtxt('test_2^28_baseline.csv', delimiter=',')
data_baseline_o3 = np.genfromtxt('test_2^28_baseline_o3.csv', delimiter=',')
x = data[0]
y = data[1]
x_o3 = data_o3[0]
y_o3 = data_o3[1]
x_baseline=data_baseline[0]
y_baseline=data_baseline[1]
x_baseline_o3=data_baseline_o3[0]
y_baseline_o3=data_baseline_o3[1]

fig, ax = plt.subplots()
# ax.set_xscale('log')
# ax.set_yscale('log')
ax.plot(x, y)
ax.plot(x_o3, y_o3,'-o')
ax.plot(x_baseline,y_baseline)
ax.plot(x_baseline_o3,y_baseline_o3)
plt.legend(("AVX2","AVX2 with -O3","Baseline","Baseline with -O3"))
plt.show()
# plt.savefig("baseline_vs_avx2.png")
```
