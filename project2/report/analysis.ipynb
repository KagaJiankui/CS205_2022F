```python
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('test_2^28_avx2.csv', delimiter=',')
data_baseline = np.genfromtxt('test_2^28_baseline.csv', delimiter=',')
x = data[0]
y = data[1]
x_baseline=data_baseline[0]
y_baseline=data_baseline[1]

fig, ax = plt.subplots()
# ax.set_xscale('log')
# ax.set_yscale('log')
ax.plot(x, y)
ax.plot(x_baseline,y_baseline)
plt.show()
plt.savefig("baseline_vs_avx2.png")
```
